```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import emukit
import GPy
from emukit.core import ParameterSpace, ContinuousParameter
from GPy.models import GPRegression
from emukit.model_wrappers import GPyModelWrapper
import itertools as it
import shap

plt.rcParams['font.sans-serif'] = ["Arial"]

def main():

    pb_data= pd.read_excel('TEP_3.xlsx',sheet_name='model_3')
    pb_data.set_axis(['Number','TEP(ul)','NMP(ul)','2-Me(ul)','Temp(℃)','MACl(%)','conc(mmol/1000ul)','NMP ratio(%)','The highest efficiency'],axis=1)
    df=pb_data[['TEP(ul)','NMP(ul)','2-Me(ul)','Temp(℃)','MACl(%)','The highest efficiency']]
    df.columns = ['TEP (μL)',  'NMP (μL)', '2-Me (μL)',  'Temp (°C)', 'MACl (%)', 'Efficiency (%)']


    TEP_min, TEP_max, TEP_step = [500, 651, 10]
    TEP_var = np.arange(TEP_min, TEP_max+TEP_step*0.1, TEP_step)
    TEP_num = len(TEP_var)
    NMP_min, NMP_max, NMP_step = [30, 81, 5]
    NMP_var = np.arange(NMP_min, NMP_max+NMP_step*0.1, NMP_step)
    NMP_num = len(NMP_var)
    TwoMe_min, TwoMe_max, TwoMe_step = [0, 50, 5]
    TwoMe_var = np.arange(TwoMe_min, TwoMe_max+TwoMe_step*0.1, TwoMe_step)
    TwoMe_num = len(TwoMe_var)
    Temp_min, Temp_max, Temp_step = [120, 150, 5]
    Temp_var = np.arange(Temp_min, Temp_max+Temp_step*0.1, Temp_step)
    Temp_num = len(Temp_var)
    MACl_min, MACl_max, MACl_step = [10, 40, 5]
    MACl_var = np.arange(MACl_min, MACl_max+MACl_step*0.1, MACl_step)
    MACl_num = len(MACl_var)
    var_array = [TEP_var, NMP_var, TwoMe_var, Temp_var, MACl_var]

    def x_normalizer(X, var_array=var_array):
        def max_min_scaler(x, x_max, x_min):
            return (x-x_min)/(x_max-x_min)
        x_norm = []
        for x in X:
            x_norm.append([max_min_scaler(x[i], max(var_array[i]), min(var_array[i])) for i in range(len(x))])
        return np.array(x_norm)


    parameter_space = ParameterSpace([
        ContinuousParameter('TEP', 0-1/(TEP_num-1)/2, 1+1/(TEP_num-1)/2),
        ContinuousParameter('NMP', 0-1/(NMP_num-1)/2, 1+1/(NMP_num-1)/2),
        ContinuousParameter('TwoMe', 0-1/(TwoMe_num-1)/2, 1+1/(TwoMe_num-1)/2),
        ContinuousParameter('Temp', 0-1/(Temp_num-1)/2, 1+1/(Temp_num-1)/2),
        ContinuousParameter('MACl', 0-1/(MACl_num-1)/2, 1+1/(MACl_num-1)/2),
    ])

    x_exp = x_normalizer(df.iloc[:,0:5].values)
    y_exp = np.transpose(([df.iloc[:,-1].values]))
    X, Y = [x_exp, y_exp]
    input_dim = 5
    ker = GPy.kern.Matern52(input_dim=input_dim, ARD=True)
    ker.lengthscale.constrain_bounded(0.01, 5)
    ker.variance.constrain_bounded(1e-2, 10000.0)
    model_gpy = GPRegression(X, -Y, ker)
    model_gpy.Gaussian_noise.variance = 0.25
    model_gpy.Gaussian_noise.variance.fix()
    model_gpy.randomize()
    model_gpy.optimize_restarts(num_restarts=30, verbose=False, messages=False)
    objective_model = GPyModelWrapper(model_gpy)
    print("lengthscale:", objective_model.model.kern.lengthscale)
    print("variance:", objective_model.model.kern.variance)


    a = np.arange(530, 681, 20)
    b = np.arange(30, 81, 10)
    c = np.arange(0, 61, 10)
    d = np.arange(120, 151, 5)
    e = np.arange(10, 41, 5)
    data = it.product(a, b, c, d, e)
    x1_grid = np.array(list(data))
    np.save('x2_grid.npy', x1_grid)
    print("x2_grid.npy")
    x1_grid_norm = np.array(x_normalizer(x1_grid))
    print("x1_grid_norm shape:", x1_grid_norm.shape)


    f_obj = lambda x: objective_model.model.predict(x)[0]
    explainer = shap.KernelExplainer(f_obj, x1_grid_norm)
    shap_values = explainer.shap_values(x1_grid_norm)
    np.save('shap2_values.npy', shap_values)
    print("shap2_values.npy")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm


arial_font = fm.FontProperties(fname=fm.findfont('Arial'))


shap_values_all = -np.load('shap2_values.npy')  # 负号还原为对原始Y的解释
x1_grid_all = np.load('x2_grid.npy')

feature_names = ['TEP', 'NMP', '2-Me', 'Temp', 'MACl']
file_suffixes = ['a', 'b', 'c', 'd', 'e']


xlims = [(520, 660), (28, 82), (-2, 52), (118, 152), (8, 42)]
xticks_list = [
    np.arange(530, 661, 20),
    np.arange(30, 86, 10),
    np.arange(0, 51, 10),
    np.arange(120, 156, 10),
    np.arange(10, 46, 10)
]


ylims = [
    (-3, 2),   # TEP
    (-1.5, 1),    # NMP
    (-1, 1.5),    # 2-Me
    (-3, 3),    # Temp
    (-7, 3.5)     # MACl
]


max_gray = 4000
if x1_grid_all.shape[0] > max_gray:
    idx = np.random.choice(x1_grid_all.shape[0], max_gray, replace=False)
    x1_grid_all_plot = x1_grid_all[idx]
    shap_values_all_plot = shap_values_all[idx]
else:
    x1_grid_all_plot = x1_grid_all
    shap_values_all_plot = shap_values_all


for i in range(5):
    name = feature_names[i]
    suffix = file_suffixes[i]
    xlim_min, xlim_max = xlims[i]
    ylim_min, ylim_max = ylims[i]
    xticks = xticks_list[i]


    x_blue = np.load(f'X{suffix}_origin_values.npy')
    shap_blue = np.load(f'shap_X{suffix}_values.npy')
    if shap_blue.ndim == 3:
        shap_blue = shap_blue.squeeze()
    shap_blue = -shap_blue  

    fig, ax = plt.subplots(figsize=(4, 3.5), dpi=200)

    ax.scatter(
        x1_grid_all_plot[:, i], shap_values_all_plot[:, i],
        color='#d3d3d3', s=20, linewidths=0
    )


    ax.scatter(
        x_blue[:, i], shap_blue[:, i],
        color='royalblue', s=20, linewidths=0
    )


    ax.set_xlabel(name, fontsize=14, fontweight='bold', fontproperties=arial_font, labelpad=12)
    ax.set_ylabel('SHAP value', fontsize=14, fontweight='bold', fontproperties=arial_font, labelpad=12)
    ax.tick_params(direction='in', width=2.5, labelsize=14, pad=8)
    ax.set_xlim(xlim_min, xlim_max)
    ax.set_xticks(xticks)
    ax.set_ylim(ylim_min, ylim_max)

    for label in ax.get_xticklabels() + ax.get_yticklabels():
        label.set_fontproperties(arial_font)
        label.set_fontweight('bold')


    for spine in ax.spines.values():
        spine.set_linewidth(2.5)

    plt.tight_layout()
    plt.show()
```
